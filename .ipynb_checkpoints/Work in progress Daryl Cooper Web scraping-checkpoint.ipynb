```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.ratemyprofessors.com/ShowRatings.jsp?tid=186093&showMyProfs=true"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html.parser')
```

```python
review_text = []
review_text_element = soup.find_all(class_="Rating__Comments-sc-1rhvpxz-1 edLmhN")

for item in review_text_element:
    review_text.append(item.text)

len(review_text)
```

```python
thumbs = []
thumbs_element = soup.find_all(class_="RatingFooter__HelpTotal-ciwspm-2 kAVFzA")

for item in thumbs_element:
    thumbs.append(item.text)

thumbs_up = []
thumbs_down = []

for i in range(0,len(thumbs)-1):
    if(i % 2 == 0):
        thumbs_up.append(thumbs[i])
    if(i % 2 == 1):
        thumbs_down.append(thumbs[i])
    
len(thumbs_up)
len(thumbs_down)
```

```python
quality = []
quality_element = soup.find_all(class_="RatingValues__RatingValue-sc-6dc747-3 kUfTQq")

for item in quality_element:
    quality.append(item.text)
```

```python
final_array = []
for text, quality_rating, tu, td in zip(review_text, quality, thumbs_up, thumbs_down):
    final_array.append({"Review Text":text, "Quality": quality_rating, "Thumbs up": tu, "Thumbs down":td})
    
df = pd.DataFrame(final_array)
df
```
